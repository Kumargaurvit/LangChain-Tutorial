#### OpenAI Embedding

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large") # Will convert text to vector of default dimensions = 3072
embeddings
```

```python
text = "This is an OpenAI Embedding tutorial"
query_result = embeddings.embed_query(text) # Need to add 5$ to openai credits to use
query_result
```

```python
len(query_result)
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024) # Custom dimension of 1024 vectors
query_result = embeddings.embed_query(text) # Need to add 5$ to openai credits to use
query_result
```
