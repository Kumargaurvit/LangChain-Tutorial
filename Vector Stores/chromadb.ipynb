#### Chroma
Chroma is a AI-native open-source vector database focused on developer productivity and happiness. Chroma is licensed under Apache 2.0.

https://python.langchain.com/v0.2/docs/integrations/vectorstores/

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_ollama import OllamaEmbeddings
```

```python
loader = TextLoader('Data/speech.txt')
docs = loader.load()
docs
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
text = text_splitter.split_documents(docs)
text
```

```python
embeddings = OllamaEmbeddings(model="llama3.2:1b")
embeddings
```

```python
vectordb = Chroma.from_documents(documents=text, embedding=embeddings)
vectordb
```

```python
# Querying
query = "What is the growth rate in India?"
```

```python
docs = vectordb.similarity_search(query, k=2)
docs[0].page_content
```

#### Saving and Loading

```python
vectordb = Chroma.from_documents(documents=text, embedding=embeddings, persist_directory="./chroma_db")
vectordb
```

```python
db2 = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
db2
```

```python
docs = db2.similarity_search(query, k=3)
docs[0].page_contentr
```

```python
retriever = db2.as_retriever()
retriever.invoke(query)[0].page_content
```
